```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline

plt.xlabel('Area (square ft)')
plt.ylabel('Price (US $)')
plt.scatter(df.Area,df.Price, color='red', marker= '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area']], df.Price)
```

The first line convert our 1-d array to 2d array with numpy reshape method.

**y = mx + b** This is the formula for the linear equation which we have studied in our school days where
m is the slope or gradient
b is the intecept

price = m * area + b

So using the slope ad intecept our linear regression model predict the price of house based on the area of it.

```python
# Convert data into 2 D array

predicted_price = reg.predict([[3300]])
print("Pridicted price for the input data is = $", predicted_price)
```

```python
%matplotlib inline

plt.xlabel('Area ', fontsize=20)
plt.ylabel('Price ', fontsize=20)
plt.scatter(df.Area, df.Price, color='red', marker='+')

# Use the correct column name 'Area'
plt.plot(df.Area, reg.predict(df[['Area']]), color='blue')

#The plt.plot(df.Area, reg.predict(df[['Area']]), color='blue') line plots the predicted prices on the current figure using a blue line.
#The x-axis represents the Area values and the y-axis represents the predicted prices.
```

**Predicting the Prices of Houses**

Now we are going to predict the prices of houses from the areas.csv file availabe on the github link https://github.com/mehmoodulhaq570/Machine-Learning-Models/tree/main/Linear%20Regression


```python
d = pd.read_csv('areas.csv')
d.head(5)
```

```python
# You can use reshape method to convert the data into 2D, as i have told previously input must be given in 2D
input_data = np.asarray(d).reshape(-1,1)
predicted_price = reg.predict(d)

# Adding a new coumn in data frame with name Prices
# Because our current predicted prices are in 2D, we convert it back to a 1D array because our areas are also in 1D array
new_df = pd.DataFrame({'Area': input_data.flatten(), 'Prices': predicted_price})
new_df
```

```python
# New input and its prediction
new_x = 3300
new_y = reg.predict([[new_x]])
# Plot data points
plt.scatter(df.Area, df.Price, color='blue')

# Plot regression line
plt.plot(df.Area, reg.predict(df[['Area']]), color='blue')

# Add crossing lines for the new point
plt.axvline(new_x, color='green')
plt.axhline(new_y, color='purple')

# Plot the new point
plt.scatter(new_x, new_y, color='orange')

# Labels and legend
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression with Crossing Lines')
```
